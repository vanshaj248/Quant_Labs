Day 1: Combinatorics and Bayes' Theorem
Key Concepts:
* Combinatorics: Counting methods - permutations P(n,r) = n!/(n-r)!, combinations C(n,r) = n!/(r!(n-r)!)
* Conditional Probability: P(A|B) = P(A∩B)/P(B)
* Bayes' Theorem: P(A|B) = P(B|A)P(A)/P(B)
* Law of Total Probability: P(B) = ΣP(B|Aᵢ)P(Aᵢ)
Today's Problems:
Finance Problem: A quantitative analyst is modeling credit default risk. Historical data shows:
* 2% of all loans default (base rate)
* A credit scoring model flags 15% of loans as "high risk"
* Among loans that actually default, 80% were flagged as "high risk"
* Among loans that don't default, 12% were flagged as "high risk"
What's the probability that a loan flagged as "high risk" will actually default? Also, if the bank approves 1000 loans and can choose any 5 for detailed review, how many ways can they select these 5 loans?
AI/ML Problem: A spam detection system processes emails with these characteristics:
* 30% of all emails are spam
* The word "offer" appears in 60% of spam emails

Methods to use:
* Bayes' Theorem: P(Default|High Risk) = P(High Risk|Default) × P(Default) / P(High Risk)
* Law of Total Probability: P(High Risk) = P(High Risk|Default)P(Default) + P(High Risk|No Default)P(No Default)
* Combinations: C(1000,5) for loan selection
* Binomial Probability: P(X=k) = C(n,k) × p^k × (1-p)^(n-k)

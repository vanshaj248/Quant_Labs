Topic Overview: Simple and Multiple Linear Regression
Simple Linear Regression models the relationship between one independent variable (X) and a dependent variable (Y): Y = β₀ + β₁X + ε
Multiple Linear Regression extends this to multiple predictors: Y = β₀ + β₁X₁ + β₂X₂ + ... + βₖXₖ + ε
Where:
* β₀ = intercept
* β₁, β₂, ..., βₖ = coefficients (slopes)
* ε = error term
Key Concepts:
* Ordinary Least Squares (OLS): Minimizes sum of squared residuals
* R-squared: Proportion of variance explained by the model
* Coefficients: Show the change in Y for a one-unit change in X
* P-values: Test statistical significance of predictors

Dat 2 Today we're covering Assumptions and Diagnostics - crucial for ensuring your regression models are valid and reliable.
Topic Overview: Regression Assumptions and Diagnostics
The 5 Key OLS Assumptions (LINE + Independence):
1. Linearity: Relationship between X and Y is linear
2. Independence: Observations are independent of each other
3. Normality: Residuals are normally distributed
4. Equal Variance (Homoskedasticity): Constant variance of residuals
5. No Perfect Multicollinearity: Predictors aren't perfectly correlated
Diagnostic Tests:
* Residual plots: Check linearity and homoskedasticity
* Q-Q plots: Test normality of residuals
* Durbin-Watson test: Check independence (autocorrelation)
* VIF (Variance Inflation Factor): Detect multicollinearity
* Cook's Distance: Identify influential outliers
Mathematical Foundation:
* Residuals: eᵢ = yᵢ - ŷᵢ
* Standardized residuals: rᵢ = eᵢ/√MSE
* VIF: VIFⱼ = 1/(1-Rⱼ²) where Rⱼ² is R² from regressing Xⱼ on other predictors

Solving Both Diagnostic Problems Step-by-Step
Problem 1: Credit Risk Model Diagnostics (Finance)
Step 1: Identify the diagnostic issues
Given problems:
* Funnel-shaped residuals vs fitted plot → Heteroskedasticity
* Q-Q plot deviates at tails → Non-normality of residuals
* VIF values > 10 → Multicollinearity
Step 2: Mathematical diagnosis
Heteroskedasticity Test (Breusch-Pagan):
* H₀: Var(εᵢ) = σ² (constant variance)
* Test statistic: LM = n × R²ₐᵤₓ
* Where R²ₐᵤₓ comes from regressing squared residuals on predictors
Sample calculation: If R²ₐᵤₓ = 0.15 with n = 1000: LM = 1000 × 0.15 = 150 Critical value χ²₀.₀₅,₄ = 9.49 Since 150 > 9.49, reject H₀ → Heteroskedasticity confirmed
VIF Calculation: For credit score (X₁), if R₁² = 0.92 (when regressed on other variables): VIF₁ = 1/(1-0.92) = 1/0.08 = 12.5
This indicates severe multicollinearity (VIF > 10).
Step 3: Remedies and implementation
For Heteroskedasticity:
1. White's robust standard errors: SE_robust(β̂ⱼ) = √[(X'X)⁻¹X'ΩX(X'X)⁻¹]ⱼⱼ where Ω = diag(e₁², e₂², ..., eₙ²)
2. Weighted Least Squares (WLS): Weight wᵢ = 1/√|eᵢ|, then minimize: Σwᵢeᵢ²
3. Log transformation of dependent variable: ln(Default_Probability) = β₀ + β₁(Credit_Score) + ...
For Multicollinearity:
1. Remove highly correlated variables: Drop debt-to-income if correlated with credit score
2. Ridge regression: β̂_ridge = (X'X + λI)⁻¹X'Y
3. Principal Components Analysis (PCA): Transform to orthogonal components
Step 4: Validation after remedies
Post-correction diagnostics:
* New VIF values should be < 5
* Breusch-Pagan test should be non-significant
* Model interpretation may change significantly

Problem 2: Customer Lifetime Value Diagnostics (AI/ML)
Step 1: Interpret diagnostic results
Given issues:
* R² = 0.78 (good fit)
* Heteroskedasticity present
* Pattern in residuals suggests non-linearity
* Durbin-Watson = 0.95 (indicates positive autocorrelation)
Step 2: Mathematical analysis of each issue
Autocorrelation Analysis: Durbin-Watson statistic: DW = Σ(eₜ - eₜ₋₁)²/Σeₜ²
Interpretation of DW = 0.95:
* DW ≈ 2: No autocorrelation
* DW < 2: Positive autocorrelation
* DW > 2: Negative autocorrelation
Since 0.95 < 1.5, we have strong positive autocorrelation.
First-order autocorrelation coefficient: ρ ≈ (2 - DW)/2 = (2 - 0.95)/2 = 0.525
Step 3: Non-linearity detection
Residual patterns suggest missing polynomial terms: Original model: CLV = β₀ + β₁(Fee) + β₂(Devices) + β₃(Hours) + β₄(Tenure)
Add quadratic terms: CLV = β₀ + β₁(Fee) + β₂(Devices) + β₃(Hours) + β₄(Tenure) + β₅(Hours²) + β₆(Tenure²)
RESET test for specification: H₀: Model is correctly specified Add ŷ², ŷ³ to model and test joint significance
Step 4: Comprehensive remedies
For Autocorrelation:
1. Cochrane-Orcutt procedure:
    * Transform: yₜ* = yₜ - ρyₜ₋₁
    * Transform: xₜ* = xₜ - ρxₜ₋₁
    * Run OLS on transformed data
2. Add lagged variables: CLV = β₀ + β₁(Fee) + ... + β₅(CLV_{t-1}) + εₜ
For Heteroskedasticity in time series:
1. GARCH modeling: Var(εₜ) = α₀ + α₁ε²_{t-1} + β₁σ²_{t-1}
2. Robust Newey-West standard errors: Accounts for both heteroskedasticity and autocorrelation
For Non-linearity:
1. Polynomial regression: Add Hours² and Tenure² terms
2. Spline regression: Piecewise polynomials for tenure effect
3. Box-Cox transformation: Find optimal λ: y^(λ) = (y^λ - 1)/λ
Step 5: Model comparison and validation
Information criteria for model selection:
* AIC: AIC = n·ln(SSE/n) + 2k
* BIC: BIC = n·ln(SSE/n) + k·ln(n)
Cross-validation for overfitting: Split data, compare out-of-sample R²
Final diagnostic check:
* Durbin-Watson should be closer to 2.0
* Residual plots should show random scatter
* VIF values < 5
* Normality tests should be non-significant

Key Mathematical Insights Applied
Diagnostic Test Statistics:
1. Breusch-Pagan: LM ~ χ²ₖ under H₀
2. Durbin-Watson: DW = 2(1-ρ) approximately
3. VIF: Measures multicollinearity severity
4. Cook's Distance: Dᵢ = (eᵢ²/k·MSE) × (hᵢᵢ/(1-hᵢᵢ)²)
Transformation Guidelines:
* Right skewed residuals: Use log(Y) or √Y
* Heteroskedasticity: Use WLS or robust SE
* Non-linearity: Add polynomial terms or use splines
* Autocorrelation: Use GLS or add lags

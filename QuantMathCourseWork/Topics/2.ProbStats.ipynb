Day 1: Combinatorics and Bayes' Theorem
Key Concepts:
* Combinatorics: Counting methods - permutations P(n,r) = n!/(n-r)!, combinations C(n,r) = n!/(r!(n-r)!)
* Conditional Probability: P(A|B) = P(A∩B)/P(B)
* Bayes' Theorem: P(A|B) = P(B|A)P(A)/P(B)
* Law of Total Probability: P(B) = ΣP(B|Aᵢ)P(Aᵢ)

```python
"""AI/ML Problem Solution"""
import math

print("SPAM DETECTION SYSTEM ANALYSIS")
print("==============================")

# Part 1: Bayes' Theorem for spam probability
P_spam = 0.30
P_ham = 0.70
P_offer_given_spam = 0.60
P_offer_given_ham = 0.05

# Calculate P("offer")
P_offer = (P_offer_given_spam * P_spam) + (P_offer_given_ham * P_ham)

# Calculate P(Spam|"offer")
P_spam_given_offer = (P_offer_given_spam * P_spam) / P_offer

print("1. PROBABILITY EMAIL IS SPAM GIVEN 'OFFER':")
print(f"P(Spam|'offer') = {P_spam_given_offer:.4f} ({P_spam_given_offer*100:.2f}%)")

# Part 2: Binomial probability for batch processing
def nCr(n, r):
    return math.comb(n, r)

def binomial_prob(n, k, p):
    return nCr(n, k) * (p ** k) * ((1 - p) ** (n - k))

n = 10
k = 3
p = 0.3

prob = binomial_prob(n, k, p)

print(f"\n2. PROBABILITY OF EXACTLY {k} SPAM EMAILS IN BATCH OF {n}:")
print(f"P(X = {k}) = C({n}, {k}) × {p}^{k} × {1-p}^{n-k}")
print(f"          = {nCr(n, k)} × {p**k:.6f} × {(1-p)**(n-k):.6f}")
print(f"          = {prob:.6f}")
print(f"          = {prob*100:.2f}%")

# Additional: Show the complete binomial distribution
print(f"\n3. COMPLETE BINOMIAL DISTRIBUTION (n={n}, p={p}):")
for i in range(n + 1):
    prob_i = binomial_prob(n, i, p)
    print(f"P(X = {i}) = {prob_i:.4f} ({prob_i*100:.1f}%)")
```

Key Concepts:
* Random Variable: A function that assigns numerical values to outcomes of random experiments
* Probability Density Function (PDF): f(x) ≥ 0, ∫f(x)dx = 1, describes probability density
* Cumulative Distribution Function (CDF): F(x) = P(X ≤ x) = ∫_{-∞}^x f(t)dt
* Discrete vs Continuous: PMF for discrete, PDF for continuous
* Properties: F'(x) = f(x), P(a < X ≤ b) = F(b) - F(a)

```python
"""Finance Problem Solution"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace

# Parameters
mu = 0  # location parameter (center)
b = 2   # scale parameter (b = 1/λ where λ = 0.5)

# 1. CDF F(x)
def laplace_cdf(x):
    """CDF of Laplace distribution"""
    if x >= 0:
        return 1 - 0.5 * np.exp(-0.5 * x)
    else:
        return 0.5 * np.exp(0.5 * x)

# 2. P(-2% < Return ≤ 3%)
prob_neg2_to_3 = laplace_cdf(3) - laplace_cdf(-2)
print(f"P(-2% < Return ≤ 3%): {prob_neg2_to_3:.6f}")

# 3. Probability of loss (negative return)
prob_loss = laplace_cdf(0)  # P(X < 0)
print(f"Probability of loss (negative return): {prob_loss:.6f}")

# 4. 95th percentile (VaR calculation)
# Solve: 1 - 0.5e^(-0.5x) = 0.95
# 0.5e^(-0.5x) = 0.05
# e^(-0.5x) = 0.1
# -0.5x = ln(0.1)
# x = -2 * ln(0.1)
var_95 = -2 * np.log(0.1)
print(f"95th percentile (VaR): {var_95:.6f}%")

# Verification using scipy
print("\nVerification using scipy.stats.laplace:")
# Using scipy's laplace distribution (scale parameter is b)
dist = laplace(loc=mu, scale=b)

# 2. P(-2% < Return ≤ 3%)
scipy_prob = dist.cdf(3) - dist.cdf(-2)
print(f"P(-2% < Return ≤ 3%): {scipy_prob:.6f}")

# 3. Probability of loss
scipy_loss = dist.cdf(0)
print(f"Probability of loss: {scipy_loss:.6f}")

# 4. 95th percentile
scipy_var_95 = dist.ppf(0.95)
print(f"95th percentile (VaR): {scipy_var_95:.6f}%")

# Visualization
plt.figure(figsize=(12, 5))

# PDF plot
plt.subplot(1, 2, 1)
x = np.linspace(-10, 10, 1000)
pdf_values = 0.5 * np.exp(-0.5 * np.abs(x))
plt.plot(x, pdf_values, 'b-', lw=2, label='PDF: f(x) = 0.5e^(-0.5|x|)')
plt.fill_between(x[(x > -2) & (x <= 3)], pdf_values[(x > -2) & (x <= 3)], 
                alpha=0.3, color='green', label='P(-2 < X ≤ 3)')
plt.axvline(x=0, color='red', linestyle='--', alpha=0.7, label='Zero return')
plt.xlabel('Daily Return (%)')
plt.ylabel('Probability Density')
plt.title('PDF of Daily Returns (Laplace Distribution)')
plt.legend()
plt.grid(True, alpha=0.3)

# CDF plot
plt.subplot(1, 2, 2)
cdf_values = [laplace_cdf(x_i) for x_i in x]
plt.plot(x, cdf_values, 'r-', lw=2, label='CDF: F(x)')
plt.axhline(y=0.95, color='purple', linestyle='--', alpha=0.7, label='95% confidence')
plt.axvline(x=var_95, color='purple', linestyle='--', alpha=0.7, label=f'95th percentile: {var_95:.2f}%')
plt.axhline(y=0.5, color='red', linestyle='--', alpha=0.7, label='50% (median)')
plt.xlabel('Daily Return (%)')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Daily Returns')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Additional calculations
print(f"\nAdditional calculations:")
print(f"Probability of return > 5%: {1 - laplace_cdf(5):.6f}")
print(f"Probability of return < -3%: {laplace_cdf(-3):.6f}")
print(f"Interquartile range (25th to 75th percentile): [{dist.ppf(0.25):.3f}%, {dist.ppf(0.75):.3f}%]")
```

* Normal Distribution: N(μ,σ²), bell-shaped, f(x) = (1/σ√2π)e^(-(x-μ)²/2σ²)
* Standard Normal: Z ~ N(0,1), used for standardization: Z = (X-μ)/σ
* Binomial Distribution: B(n,p), counts successes in n trials, P(X=k) = C(n,k)p^k(1-p)^(n-k)
* Poisson Distribution: Pois(λ), counts events in time/space, P(X=k) = (λ^k e^(-λ))/k!

Day 5: Law of Large Numbers and Central Limit Theorem
Key Concepts:
* Law of Large Numbers (LLN): As n→∞, sample mean X̄ₙ converges to population mean μ
* Weak LLN: X̄ₙ converges in probability: P(|X̄ₙ - μ| > ε) → 0
* Strong LLN: X̄ₙ converges almost surely: P(X̄ₙ → μ) = 1
* Central Limit Theorem (CLT): For large n, X̄ₙ ~ N(μ, σ²/n) regardless of original distribution
* Standardized form: Z = (X̄ₙ - μ)/(σ/√n) ~ N(0,1)

```python
"""Finance Problem Solution"""
import numpy as np
from scipy import stats

# Given parameters
mu = 0.8  # %
sigma = 3.2  # %

print("Hedge Fund Monte Carlo Simulation Analysis")
print("=" * 50)
print(f"Individual trade: μ = {mu}%, σ = {sigma}%")
print()

# 1. Probability that average return over 100 trades is between 0.5% and 1.1%
n1 = 100
se1 = sigma / np.sqrt(n1)
z_low1 = (0.5 - mu) / se1
z_high1 = (1.1 - mu) / se1
prob1 = stats.norm.cdf(z_high1) - stats.norm.cdf(z_low1)

print("1. Probability that average return over 100 trades is between 0.5% and 1.1%:")
print(f"   Standard Error (SE) = σ/√n = {sigma}/√{n1} = {se1:.4f}%")
print(f"   Z-score for 0.5%: ({0.5} - {mu})/{se1:.4f} = {z_low1:.4f}")
print(f"   Z-score for 1.1%: ({1.1} - {mu})/{se1:.4f} = {z_high1:.4f}")
print(f"   P({z_low1:.4f} ≤ Z ≤ {z_high1:.4f}) = {prob1:.4f}")
print(f"   Answer: {prob1:.4f} or {prob1*100:.2f}%")
print()

# 2. Number of trades needed for sample mean within ±0.2% with 95% probability
margin = 0.2
z_95 = stats.norm.ppf(0.975)  # 1.96 for 95% CI
n2 = (z_95 * sigma / margin) ** 2
n2 = np.ceil(n2)  # Round up to nearest integer

print("2. Number of trades needed for sample mean within ±0.2% with 95% probability:")
print(f"   Formula: n ≥ (z_{0.95} × σ / margin)²")
print(f"   n ≥ ({z_95:.4f} × {sigma} / {margin})²")
print(f"   n ≥ ({z_95 * sigma / margin:.4f})²")
print(f"   n ≥ {n2:.0f} trades")
print(f"   Answer: {n2:.0f} trades")
print()

# 3. 95% confidence interval for average return over 250 trades
n3 = 250
se3 = sigma / np.sqrt(n3)
ci_low = mu - z_95 * se3
ci_high = mu + z_95 * se3

print("3. 95% confidence interval for average return over 250 trades:")
print(f"   Standard Error = {sigma}/√{n3} = {se3:.4f}%")
print(f"   95% CI = {mu}% ± {z_95:.4f} × {se3:.4f}%")
print(f"   = {mu}% ± {z_95 * se3:.4f}%")
print(f"   = ({ci_low:.4f}%, {ci_high:.4f}%)")
print(f"   Answer: ({ci_low:.4f}%, {ci_high:.4f}%)")
print()

# 4. Expected number of simulations with average returns above 1.2%
n4 = 100  # trades per simulation
num_simulations = 1000
se4 = sigma / np.sqrt(n4)
z_target = (1.2 - mu) / se4
prob_above = 1 - stats.norm.cdf(z_target)
expected_count = num_simulations * prob_above

print("4. Expected number of simulations with average returns above 1.2%:")
print(f"   Standard Error for 100 trades = {sigma}/√{n4} = {se4:.4f}%")
print(f"   Z-score for 1.2%: ({1.2} - {mu})/{se4:.4f} = {z_target:.4f}")
print(f"   P(X̄ > 1.2%) = P(Z > {z_target:.4f}) = {prob_above:.4f}")
print(f"   Expected count = {num_simulations} × {prob_above:.4f} = {expected_count:.1f}")
print(f"   Answer: Approximately {expected_count:.0f} simulations")
print()

# Additional analysis and visualization
print("Additional Analysis:")
print("=" * 50)

# Show how standard error decreases with sample size
sample_sizes = [10, 25, 50, 100, 250, 500, 1000]
print("Standard Error for different sample sizes:")
for n in sample_sizes:
    se = sigma / np.sqrt(n)
    print(f"   n = {n:4d}: SE = {se:.4f}%")

# Monte Carlo simulation to verify results
np.random.seed(42)
print(f"\nMonte Carlo simulation verification ({num_simulations} simulations of {n4} trades each):")

# Simulate 1000 samples of 100 trades each
# Using normal distribution (CLT assumption)
simulated_means = []
for _ in range(num_simulations):
    sample = np.random.normal(mu, sigma, n4)
    simulated_means.append(np.mean(sample))

simulated_means = np.array(simulated_means)

# Verify probability from part 1
in_range = np.sum((simulated_means >= 0.5) & (simulated_means <= 1.1))
print(f"   P(0.5% ≤ X̄ ≤ 1.1%): simulated = {in_range/num_simulations:.4f}, theoretical = {prob1:.4f}")

# Verify probability from part 4
above_1_2 = np.sum(simulated_means > 1.2)
print(f"   P(X̄ > 1.2%): simulated = {above_1_2/num_simulations:.4f}, theoretical = {prob_above:.4f}")

# Show distribution of sample means
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(simulated_means, bins=30, alpha=0.7, density=True, label='Simulated means')

# Overlay theoretical normal distribution
x = np.linspace(mu - 3*se4, mu + 3*se4, 100)
plt.plot(x, stats.norm.pdf(x, mu, se4), 'r-', lw=2, label='Theoretical N(μ, σ²/n)')

plt.axvline(x=mu, color='green', linestyle='--', label=f'True mean (μ = {mu}%)')
plt.axvline(x=0.5, color='blue', linestyle='--', alpha=0.7, label='0.5% threshold')
plt.axvline(x=1.1, color='pink', linestyle='--', alpha=0.7, label='1.1% threshold')
plt.axvline(x=1.2, color='red', linestyle='--', alpha=0.7, label='1.2% threshold')

plt.xlabel('Average Return (%)')
plt.ylabel('Density')
plt.title(f'Distribution of Sample Means (n = {n4} trades)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Confidence interval coverage verification
print(f"\nConfidence interval coverage verification:")
ci_low_sim = np.percentile(simulated_means, 2.5)
ci_high_sim = np.percentile(simulated_means, 97.5)
print(f"   95% empirical CI from simulations: ({ci_low_sim:.4f}%, {ci_high_sim:.4f}%)")
print(f"   Theoretical 95% CI: ({mu - 1.96*se4:.4f}%, {mu + 1.96*se4:.4f}%)")

within_ci = np.sum((simulated_means >= mu - 1.96*se4) & (simulated_means <= mu + 1.96*se4))
print(f"   Percentage of means within theoretical 95% CI: {within_ci/num_simulations*100:.2f}%")
```

```python
"""AI/ML Problem Solution"""
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Given parameters
mu = 0.0      # True mean error (under null hypothesis)
sigma = 0.25  # Standard deviation of prediction errors
true_mu = 0.02  # Actual mean error (biased model)

print("AI/ML Model Performance Analysis")
print("=" * 50)
print(f"Individual prediction errors: μ = {mu}, σ = {sigma}")
print(f"True model bias: μ = {true_mu}")
print()

# 1. Probability that mean error over 400 predictions is between -0.05 and 0.05
n1 = 400
se1 = sigma / np.sqrt(n1)
z_low1 = (-0.05 - mu) / se1
z_high1 = (0.05 - mu) / se1
prob1 = stats.norm.cdf(z_high1) - stats.norm.cdf(z_low1)

print("1. Probability that mean error over 400 predictions is between -0.05 and 0.05:")
print(f"   Standard Error (SE) = σ/√n = {sigma}/√{n1} = {se1:.6f}")
print(f"   Z-score for -0.05: ({-0.05} - {mu})/{se1:.6f} = {z_low1:.4f}")
print(f"   Z-score for 0.05: ({0.05} - {mu})/{se1:.6f} = {z_high1:.4f}")
print(f"   P({z_low1:.4f} ≤ Z ≤ {z_high1:.4f}) = {prob1:.6f}")
print(f"   Answer: {prob1:.6f} or {prob1*100:.4f}%")
print()

# 2. Sample size needed for |X̄ - 0| < 0.03 with probability 0.99
margin = 0.03
confidence = 0.99
z_critical = stats.norm.ppf((1 + confidence) / 2)  # Two-tailed
n2 = (z_critical * sigma / margin) ** 2
n2 = np.ceil(n2)  # Round up to nearest integer

print("2. Sample size needed for |X̄ - 0| < 0.03 with probability 0.99:")
print(f"   Confidence level: {confidence*100}%")
print(f"   Margin of error: ±{margin}")
print(f"   z-critical value: {z_critical:.4f}")
print(f"   Formula: n ≥ (z × σ / margin)²")
print(f"   n ≥ ({z_critical:.4f} × {sigma} / {margin})²")
print(f"   n ≥ {n2:.0f} predictions")
print(f"   Answer: {n2:.0f} predictions")
print()

# 3. 90% confidence interval for true mean error based on 900 predictions
n3 = 900
confidence_level = 0.90
z_90 = stats.norm.ppf((1 + confidence_level) / 2)  # 1.645 for 90% CI
se3 = sigma / np.sqrt(n3)
ci_low = mu - z_90 * se3
ci_high = mu + z_90 * se3

print("3. 90% confidence interval for true mean error based on 900 predictions:")
print(f"   Standard Error = {sigma}/√{n3} = {se3:.6f}")
print(f"   z-value for 90% CI: {z_90:.4f}")
print(f"   90% CI = {mu} ± {z_90:.4f} × {se3:.6f}")
print(f"   = {mu} ± {z_90 * se3:.6f}")
print(f"   = ({ci_low:.6f}, {ci_high:.6f})")
print(f"   Answer: ({ci_low:.6f}, {ci_high:.6f})")
print()

# 4. Probability of detecting bias (μ = 0.02) with 625 predictions using α = 0.05
n4 = 625
alpha = 0.05
se4 = sigma / np.sqrt(n4)

# Critical values for two-tailed test
z_alpha = stats.norm.ppf(1 - alpha/2)  # 1.96 for α = 0.05
critical_value = z_alpha * se4

print("4. Probability of detecting bias with 625 predictions (α = 0.05):")
print(f"   Standard Error = {sigma}/√{n4} = {se4:.6f}")
print(f"   Critical value for α = {alpha}: ±{critical_value:.6f}")
print(f"   Reject H₀ if |X̄ - 0| > {critical_value:.6f}")

# Since true μ = 0.02 > 0, we focus on the right tail
# Probability of detection = P(X̄ > critical_value | μ = true_mu)
z_detect = (critical_value - true_mu) / se4
power = 1 - stats.norm.cdf(z_detect)

print(f"   True mean: μ = {true_mu}")
print(f"   Z-score for detection: ({critical_value:.6f} - {true_mu})/{se4:.6f} = {z_detect:.4f}")
print(f"   Power = P(X̄ > {critical_value:.6f} | μ = {true_mu}) = P(Z > {z_detect:.4f}) = {power:.6f}")
print(f"   Answer: {power:.6f} or {power*100:.4f}%")
print()

# Additional analysis
print("Additional Analysis:")
print("=" * 50)

# Show how power changes with sample size
sample_sizes = [100, 225, 400, 625, 900, 1225, 1600]
print("Power analysis for detecting bias μ = 0.02 with α = 0.05:")
for n in sample_sizes:
    se = sigma / np.sqrt(n)
    crit_val = stats.norm.ppf(1 - alpha/2) * se
    z_power = (crit_val - true_mu) / se
    power_val = 1 - stats.norm.cdf(z_power)
    print(f"   n = {n:4d}: Power = {power_val:.4f}")

# Monte Carlo simulation to verify results
np.random.seed(42)
print(f"\nMonte Carlo simulation verification:")

# Simulate the detection probability
n_simulations = 10000
detected_count = 0

for _ in range(n_simulations):
    # Generate sample from true distribution (μ = 0.02)
    sample = np.random.normal(true_mu, sigma, n4)
    sample_mean = np.mean(sample)
    
    # Check if we would reject H₀: μ = 0
    if abs(sample_mean - mu) > critical_value:
        detected_count += 1

empirical_power = detected_count / n_simulations
print(f"   Empirical detection probability: {empirical_power:.6f}")
print(f"   Theoretical power: {power:.6f}")

# Visualization
plt.figure(figsize=(12, 5))

# Distribution of sample means under null and alternative hypotheses
x = np.linspace(-0.05, 0.05, 1000)
null_dist = stats.norm.pdf(x, mu, se4)
alt_dist = stats.norm.pdf(x, true_mu, se4)

plt.subplot(1, 2, 1)
plt.plot(x, null_dist, 'b-', label='Under H₀: μ = 0')
plt.plot(x, alt_dist, 'r-', label='Under H₁: μ = 0.02')
plt.axvline(x=critical_value, color='green', linestyle='--', label=f'Critical value: {critical_value:.4f}')
plt.axvline(x=-critical_value, color='green', linestyle='--')
plt.fill_between(x[x > critical_value], alt_dist[x > critical_value], alpha=0.3, color='red', label='Power')
plt.fill_between(x[x < -critical_value], alt_dist[x < -critical_value], alpha=0.3, color='red')
plt.xlabel('Sample Mean Error')
plt.ylabel('Density')
plt.title(f'Distribution of Sample Means (n = {n4})')
plt.legend()
plt.grid(True, alpha=0.3)

# Power curve as function of sample size
sample_sizes_range = np.arange(100, 2001, 50)
powers = []
for n in sample_sizes_range:
    se = sigma / np.sqrt(n)
    crit_val = stats.norm.ppf(1 - alpha/2) * se
    z_pwr = (crit_val - true_mu) / se
    powers.append(1 - stats.norm.cdf(z_pwr))

plt.subplot(1, 2, 2)
plt.plot(sample_sizes_range, powers, 'b-')
plt.axhline(y=0.8, color='red', linestyle='--', alpha=0.7, label='80% power')
plt.axhline(y=0.9, color='green', linestyle='--', alpha=0.7, label='90% power')
plt.axvline(x=n2, color='purple', linestyle='--', alpha=0.7, label=f'n for 99% CI: {n2:.0f}')
plt.xlabel('Sample Size (n)')
plt.ylabel('Power')
plt.title('Power Curve for Detecting Bias μ = 0.02 (α = 0.05)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Confidence interval demonstration
print(f"\nConfidence interval demonstration with n = {n3}:")
print(f"   If we repeatedly sample 900 predictions, 90% of the CIs will contain the true mean")
print(f"   When true μ = 0, CI should contain 0 about 90% of the time")
print(f"   When true μ = {true_mu}, CI should contain {true_mu} about 90% of the time")

# Simulate CI coverage
np.random.seed(42)
n_ci_sim = 1000
ci_contains_true = 0
ci_contains_null = 0

for _ in range(n_ci_sim):
    sample = np.random.normal(true_mu, sigma, n3)
    sample_mean = np.mean(sample)
    ci_low_sim = sample_mean - z_90 * se3
    ci_high_sim = sample_mean + z_90 * se3
    
    if ci_low_sim <= true_mu <= ci_high_sim:
        ci_contains_true += 1
    if ci_low_sim <= mu <= ci_high_sim:
        ci_contains_null += 1

print(f"   CI coverage for true μ = {true_mu}: {ci_contains_true/n_ci_sim*100:.2f}%")
print(f"   CI coverage for null μ = 0: {ci_contains_null/n_ci_sim*100:.2f}%")
```

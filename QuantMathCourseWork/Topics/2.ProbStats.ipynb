Day 1: Combinatorics and Bayes' Theorem
Key Concepts:
* Combinatorics: Counting methods - permutations P(n,r) = n!/(n-r)!, combinations C(n,r) = n!/(r!(n-r)!)
* Conditional Probability: P(A|B) = P(A∩B)/P(B)
* Bayes' Theorem: P(A|B) = P(B|A)P(A)/P(B)
* Law of Total Probability: P(B) = ΣP(B|Aᵢ)P(Aᵢ)

```python
"""AI/ML Problem Solution"""
import math

print("SPAM DETECTION SYSTEM ANALYSIS")
print("==============================")

# Part 1: Bayes' Theorem for spam probability
P_spam = 0.30
P_ham = 0.70
P_offer_given_spam = 0.60
P_offer_given_ham = 0.05

# Calculate P("offer")
P_offer = (P_offer_given_spam * P_spam) + (P_offer_given_ham * P_ham)

# Calculate P(Spam|"offer")
P_spam_given_offer = (P_offer_given_spam * P_spam) / P_offer

print("1. PROBABILITY EMAIL IS SPAM GIVEN 'OFFER':")
print(f"P(Spam|'offer') = {P_spam_given_offer:.4f} ({P_spam_given_offer*100:.2f}%)")

# Part 2: Binomial probability for batch processing
def nCr(n, r):
    return math.comb(n, r)

def binomial_prob(n, k, p):
    return nCr(n, k) * (p ** k) * ((1 - p) ** (n - k))

n = 10
k = 3
p = 0.3

prob = binomial_prob(n, k, p)

print(f"\n2. PROBABILITY OF EXACTLY {k} SPAM EMAILS IN BATCH OF {n}:")
print(f"P(X = {k}) = C({n}, {k}) × {p}^{k} × {1-p}^{n-k}")
print(f"          = {nCr(n, k)} × {p**k:.6f} × {(1-p)**(n-k):.6f}")
print(f"          = {prob:.6f}")
print(f"          = {prob*100:.2f}%")

# Additional: Show the complete binomial distribution
print(f"\n3. COMPLETE BINOMIAL DISTRIBUTION (n={n}, p={p}):")
for i in range(n + 1):
    prob_i = binomial_prob(n, i, p)
    print(f"P(X = {i}) = {prob_i:.4f} ({prob_i*100:.1f}%)")
```

Key Concepts:
* Random Variable: A function that assigns numerical values to outcomes of random experiments
* Probability Density Function (PDF): f(x) ≥ 0, ∫f(x)dx = 1, describes probability density
* Cumulative Distribution Function (CDF): F(x) = P(X ≤ x) = ∫_{-∞}^x f(t)dt
* Discrete vs Continuous: PMF for discrete, PDF for continuous
* Properties: F'(x) = f(x), P(a < X ≤ b) = F(b) - F(a)

```python
"""Finance Problem Solution"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace

# Parameters
mu = 0  # location parameter (center)
b = 2   # scale parameter (b = 1/λ where λ = 0.5)

# 1. CDF F(x)
def laplace_cdf(x):
    """CDF of Laplace distribution"""
    if x >= 0:
        return 1 - 0.5 * np.exp(-0.5 * x)
    else:
        return 0.5 * np.exp(0.5 * x)

# 2. P(-2% < Return ≤ 3%)
prob_neg2_to_3 = laplace_cdf(3) - laplace_cdf(-2)
print(f"P(-2% < Return ≤ 3%): {prob_neg2_to_3:.6f}")

# 3. Probability of loss (negative return)
prob_loss = laplace_cdf(0)  # P(X < 0)
print(f"Probability of loss (negative return): {prob_loss:.6f}")

# 4. 95th percentile (VaR calculation)
# Solve: 1 - 0.5e^(-0.5x) = 0.95
# 0.5e^(-0.5x) = 0.05
# e^(-0.5x) = 0.1
# -0.5x = ln(0.1)
# x = -2 * ln(0.1)
var_95 = -2 * np.log(0.1)
print(f"95th percentile (VaR): {var_95:.6f}%")

# Verification using scipy
print("\nVerification using scipy.stats.laplace:")
# Using scipy's laplace distribution (scale parameter is b)
dist = laplace(loc=mu, scale=b)

# 2. P(-2% < Return ≤ 3%)
scipy_prob = dist.cdf(3) - dist.cdf(-2)
print(f"P(-2% < Return ≤ 3%): {scipy_prob:.6f}")

# 3. Probability of loss
scipy_loss = dist.cdf(0)
print(f"Probability of loss: {scipy_loss:.6f}")

# 4. 95th percentile
scipy_var_95 = dist.ppf(0.95)
print(f"95th percentile (VaR): {scipy_var_95:.6f}%")

# Visualization
plt.figure(figsize=(12, 5))

# PDF plot
plt.subplot(1, 2, 1)
x = np.linspace(-10, 10, 1000)
pdf_values = 0.5 * np.exp(-0.5 * np.abs(x))
plt.plot(x, pdf_values, 'b-', lw=2, label='PDF: f(x) = 0.5e^(-0.5|x|)')
plt.fill_between(x[(x > -2) & (x <= 3)], pdf_values[(x > -2) & (x <= 3)], 
                alpha=0.3, color='green', label='P(-2 < X ≤ 3)')
plt.axvline(x=0, color='red', linestyle='--', alpha=0.7, label='Zero return')
plt.xlabel('Daily Return (%)')
plt.ylabel('Probability Density')
plt.title('PDF of Daily Returns (Laplace Distribution)')
plt.legend()
plt.grid(True, alpha=0.3)

# CDF plot
plt.subplot(1, 2, 2)
cdf_values = [laplace_cdf(x_i) for x_i in x]
plt.plot(x, cdf_values, 'r-', lw=2, label='CDF: F(x)')
plt.axhline(y=0.95, color='purple', linestyle='--', alpha=0.7, label='95% confidence')
plt.axvline(x=var_95, color='purple', linestyle='--', alpha=0.7, label=f'95th percentile: {var_95:.2f}%')
plt.axhline(y=0.5, color='red', linestyle='--', alpha=0.7, label='50% (median)')
plt.xlabel('Daily Return (%)')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Daily Returns')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Additional calculations
print(f"\nAdditional calculations:")
print(f"Probability of return > 5%: {1 - laplace_cdf(5):.6f}")
print(f"Probability of return < -3%: {laplace_cdf(-3):.6f}")
print(f"Interquartile range (25th to 75th percentile): [{dist.ppf(0.25):.3f}%, {dist.ppf(0.75):.3f}%]")
```

Key Concepts:
* Normal Distribution: N(μ,σ²), bell-shaped, f(x) = (1/σ√2π)e^(-(x-μ)²/2σ²)
* Standard Normal: Z ~ N(0,1), used for standardization: Z = (X-μ)/σ
* Binomial Distribution: B(n,p), counts successes in n trials, P(X=k) = C(n,k)p^k(1-p)^(n-k)
* Poisson Distribution: Pois(λ), counts events in time/space, P(X=k) = (λ^k e^(-λ))/k!

Day 5: Law of Large Numbers and Central Limit Theorem
Key Concepts:
* Law of Large Numbers (LLN): As n→∞, sample mean X̄ₙ converges to population mean μ
* Weak LLN: X̄ₙ converges in probability: P(|X̄ₙ - μ| > ε) → 0
* Strong LLN: X̄ₙ converges almost surely: P(X̄ₙ → μ) = 1
* Central Limit Theorem (CLT): For large n, X̄ₙ ~ N(μ, σ²/n) regardless of original distribution
* Standardized form: Z = (X̄ₙ - μ)/(σ/√n) ~ N(0,1)

Finance Problem Solution: Trading Strategy Analysis
Given: Individual trades: μ = 0.8%, σ = 3.2%
Part 1: Probability that average return over 100 trades is between 0.5% and 1.1%
By CLT: X̄₁₀₀ ~ N(μ, σ²/n) = N(0.8, 3.2²/100) = N(0.8, 0.1024) Standard error: σ_X̄ = σ/√n = 3.2/√100 = 0.32
Standardize:
Z₁ = (0.5 - 0.8)/0.32 = -0.3/0.32 = -0.9375
Z₂ = (1.1 - 0.8)/0.32 = 0.3/0.32 = 0.9375
Using standard normal table:
P(0.5% < X̄ < 1.1%) = Φ(0.9375) - Φ(-0.9375)
                    ≈ 0.826 - 0.174 = 0.652 or 65.2%
Part 2: Sample size for X̄ within ±0.2% of true mean with 95% probability
We want: P(|X̄ - μ| ≤ 0.2) = 0.95 This means: P(|Z| ≤ 0.2/(σ/√n)) = 0.95 So: 0.2/(σ/√n) = 1.96 (95% confidence)
0.2/(3.2/√n) = 1.96
0.2√n/3.2 = 1.96
√n = 1.96 × 3.2/0.2 = 31.36
n = (31.36)² ≈ 983 trades
Part 3: 95% confidence interval for average return over 250 trades
Standard error: σ_X̄ = 3.2/√250 = 3.2/15.81 = 0.202
95% CI: X̄ ± 1.96 × σ_X̄ = μ ± 1.96 × 0.202
95% CI = 0.8% ± 1.96 × 0.202% = 0.8% ± 0.396%
95% CI = [0.404%, 1.196%]
Part 4: Number of simulations with average return > 1.2% (out of 1000)
For 100 trades: X̄ ~ N(0.8, 0.32²)

AI/ML Problem Solution: Model Performance Analysis

Part 1: Probability that mean error over 400 predictions is between -0.05 and 0.05
By CLT: X̄₄₀₀ ~ N(0, 0.25²/400) = N(0, 0.000156) Standard error: σ_X̄ = 0.25/√400 = 0.25/20 = 0.0125
Standardize:

We want: P(|X̄| ≤ 0.03) = 0.99 This means: P(|Z| ≤ 0.03/(σ/√n)) = 0.99 From normal table: Φ(2.576) = 0.995, so for 99%: z = 2.576

Part 4: Power analysis for detecting bias μ = 0.02 with n = 625, α = 0.05
Null hypothesis: H₀: μ = 0 Alternative hypothesis: H₁: μ ≠ 0 (two-tailed test)
Under H₀: X̄ ~ N(0, 0.25²/625) Standard error: σ_X̄ = 0.25/√625 = 0.25/25 = 0.01
Critical values for α = 0.05 (two-tailed):
Critical values = ±1.96 × 0.01 = ±0.0196
Reject H₀ if |X̄| > 0.0196
Under H₁ (true μ = 0.02): X̄ ~ N(0.02, 0.01²)
Power = P(Reject H₀ | μ = 0.02):
Power = P(X̄ > 0.0196 | μ = 0.02) + P(X̄ < -0.0196 | μ = 0.02)

Z₁ = (0.0196 - 0.02)/0.01 = -0.04/0.01 = -0.4
Z₂ = (-0.0196 - 0.02)/0.01 = -0.0396/0.01 = -3.96

Power = P(Z > -0.4) + P(Z < -3.96)
Power ≈ 0.655 + 0.00004 ≈ 0.655 or 65.5%
